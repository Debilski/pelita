```python
from pelita import game, layout
```

```python
l = """
############
#  .    .  #
#   #      #
#  2  3    #
#  0.  1   #
#   #  .   #
############
"""
```

```python
loaded_layout = layout.parse_layout(l)
```

```python
from random import choice

def p1(bot, state):
    return bot.position, state
def p2(bot, state):
    return choice(bot.legal_positions), state
```

```python
game.run_game([p1, p2], rounds=20, layout_dict=loaded_layout)
```

```python
game.run_game([p1, p2], rounds=20, layout_dict=loaded_layout)
```

```python
state = game.setup_game([p1, p2], max_rounds=20, layout_dict=loaded_layout)
while not state.get('gameover'):
    state = game.play_turn(state)
    print(game.GameState(**state).pretty_str())
```
